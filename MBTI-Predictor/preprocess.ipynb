```python
import glob
import json
import pandas as pd
import re

multiplier = 0.0
layer = 13
json_files = glob.glob(f'caa_datasets_16_2.5Lite/results_layer={layer}_multiplier={multiplier}_behavior=caa_datasets_16_type=open_ended_use_base_model=False_model_size=7b.json')

all_data = []
# 正則表達式用於從 'question' 中提取四個大寫字母組成的 MBTI 類型
mbti_regex = re.compile(r"You are a ([A-Z]{4})")

for file_name in json_files:
    try:
        with open(file_name, 'r', encoding='utf-8') as f:
            data = json.load(f)

        # 處理 JSON 檔案中的每個項目
        if isinstance(data, list):
            for item in data:
                question = item.get('question', '')
                model_output = item.get('model_output', '')

                # 從 question 欄位中提取 MBTI 類型
                mbti_match = mbti_regex.search(question)
                mbti_type = mbti_match.group(1) if mbti_match else 'Unknown'

                all_data.append({
                    'MBTI_Type': mbti_type,
                    'Response': model_output
                })
        # 忽略非列表格式的 JSON 檔案

    except (json.JSONDecodeError, Exception) as e:
        # 跳過無效的 JSON 檔案或處理錯誤
        pass

df = pd.DataFrame(all_data)

output_file = f'caa_{layer}_{multiplier}.csv'
df.to_csv(output_file, index=False)

print(f"所有 JSON 檔案的內容已整理並儲存到 {output_file} 中，總共處理了 {len(df)} 筆資料。")
```
